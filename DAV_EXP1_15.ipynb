**WORKING WITH DPLYR**

```python
install.packages("dplyr")
```

```python

sales_data <- data.frame(
  product = c("A", "B", "A", "C", "B", "C", "A"),
  quantity = c(100, 150, 120, 80, 200, 90, 110),
  price = c(10, 20, 15, 25, 18, 30, 12)
)

# Load the dplyr library
library(dplyr)

# Displaying the original data frame
print("Original Sales Data:")
print(sales_data)
```

```python
#  Select specific columns
selected_columns <- select(sales_data, product, quantity)
print("Selected Columns (Product and Quantity):")
print(selected_columns)
```

```python
#  Filter rows based on a condition (e.g., quantity greater than 100)
filtered_data <- filter(sales_data, quantity > 100)
print("Filtered Data (Quantity > 100):")
print(filtered_data)
```

```python

#  Arrange data based on a column (e.g., product in ascending order)
arranged_data <- arrange(sales_data, product)
print("Arranged Data (Product in Ascending Order):")
print(arranged_data)
```

```python
# Mutate - Create a new calculated column (e.g., total sales)
mutated_data <- mutate(sales_data, total_sales = quantity * price)
print("Mutated Data (Added Total Sales Column):")
print(mutated_data)
```

```python
#  Summarize data (e.g., calculate the total quantity sold)
summary_stats <- summarize(sales_data, total_quantity = sum(quantity))
print("Summary Statistics (Total Quantity Sold):")
print(summary_stats)
```

**WORKING WITH ggplot2**

```python
install.packages("GGPLOT2")
```

```python
# Load the ggplot2 library
library(ggplot2)
```

```python
#Scatter plot: Quantity vs. Price
scatter_plot <- ggplot(sales_data, aes(x = price, y = quantity)) +
  geom_point() +
  labs(title = "Scatter Plot of Quantity vs. Price", x = "Price", y = "Quantity")

print(scatter_plot)
```

```python
# Bar chart: Total quantity sold per product
bar_chart <- ggplot(sales_data, aes(x = product, y = quantity)) +
  geom_bar(stat = "sum", fill = "skyblue") +
  labs(title = "Total Quantity Sold per Product", x = "Product", y = "Total Quantity")


print(bar_chart)
```

```python
#Line chart: Total sales over time (assuming a time variable is available)
# In this case, let's use a sequential number as a proxy for time
sales_data$time <- seq_len(nrow(sales_data))

line_chart <- ggplot(sales_data, aes(x = time, y = quantity * price, group = 1)) +
  geom_line(color = "red") +
  labs(title = "Total Sales Over Time", x = "Time", y = "Total Sales")


print(line_chart)
```
